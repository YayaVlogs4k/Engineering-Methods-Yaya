### a. Root Finding
Use Newton-Raphson method to find root of f(x) = x^3 - x - 2 with the initial guess x₀ = 1.5 and tolerance = 1e-5.

```python
def f(x):
    return (x**3)-x-2
def df(x):
    return 3*(x**2)-1
x0=1.5
tolerance=1e-5
x_n=x0
for i in range(100):
    f_xn=f(x_n)
    df_xn=df(x_n)
    if df_xn == 0:
        print("Derivative is zero. No solution found.")
        break
    x_n1 = x_n - f_xn / df_xn
    if abs(f(x_n1)) < tolerance:
        break
    x_n = x_n1
print("Approximate root using Newton-Raphson Method:", x_n1)
```

### b. Numerical Differentiation
Use central difference to estimate the first and second derivatives of y =  1/(x^2 + 3x + 5) at x = 5.

```python
x=5
h=0.0001
def f(x):
    return 1/(x**2+3*x+5)
def df(x):
    return (f(x+h)-f(x-h))/(2*h)
def d2f(x):
    return (f(x+h) - 2*f(x) + f(x-h)) / (h**2)
first = df(5)
second = d2f(5)
print(first, second)
```
