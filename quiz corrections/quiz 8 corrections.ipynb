```python
import numpy as np
import math
# aug is the upper-triangular augmented matrix [A|b]
n = aug.shape[0]
x = np.zeros(n)

for i in range(n - 1, -1, -1):          # from last row to first
    s = 0.0
    for j in range(i + 1, n):            # sum A[i,j] * x[j] for known x[j]
        s += aug[i, j] * x[j]
    pivot = aug[i, i]
    if pivot == 0:
        raise ValueError(f"Zero pivot at row {i}; system is singular or needs pivoting.")
    x[i] = (aug[i, -1] - s) / pivot

print("Solution x =", x)
```
