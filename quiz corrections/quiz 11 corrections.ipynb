```python
#number 1 answer
import numpy as np

# define function
def f(x):
    return np.exp(-x**2)

a, b = 0, 1
n = 10
x = np.linspace(a, b, n+1)
h = (b - a) / n

trap_integral = 0
for i in range(n):
    trap_integral += 0.5 * h * (f(x[i]) + f(x[i+1]))

print(f"Trapezoidal Rule Approximation: {trap_integral:.6f}")
```

```python
#number 2 answer
# eulers method
import numpy as np

def f(x):
    return np.exp(-x**2)

h = 0.1
num_steps = 10
x0 = 0
y0 = 1
xs = [x0]
ys = [y0]

for i in range(num_steps):
    x_new = xs[-1] + h
    y_new = ys[-1] + h * f(xs[-1])
    xs.append(x_new)
    ys.append(y_new)

print(xs)
print(ys)
```
